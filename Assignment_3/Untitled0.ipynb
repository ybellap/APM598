```python
import nbformat
from nbformat.v4 import new_notebook, new_markdown_cell

# Names of the files to be combined
notebooks = ['HW3_exe1.ipynb', 'hw3_exe2.ipynb', 'HW3_exe3.ipynb']

# Create a new notebook
combined_notebook = new_notebook()

# Function to read and normalize a notebook
def read_and_normalize_notebook(path):
    with open(path, 'r', encoding='utf-8') as f:
        nb = nbformat.read(f, as_version=4)
    # Normalize the notebook to ensure each cell has a unique ID
    nbformat.validate(nb, version=4)  # Validation is recommended after normalization
    return nb

# Loop through the notebook files and combine them
for nb_path in notebooks:
    current_notebook = read_and_normalize_notebook(nb_path)
    # Optionally, add a markdown cell with the filename as a separator
    combined_notebook.cells.append(new_markdown_cell(f"# {nb_path}"))
    combined_notebook.cells.extend(current_notebook.cells)

# Save the combined notebook
with open('combined_notebook.ipynb', 'w', encoding='utf-8') as f:
    nbformat.write(combined_notebook, f)

print("Notebooks combined into combined_notebook.ipynb")
```
